```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.collections import PatchCollection
from matplotlib.patches import Arc
import random
```

```python
def fibonacci(n=10):
    a, b = 0,1
    for i in range(n):
        c = a + b
        yield (c,b)
        a,b=b,c
```

```python
def fourth_series(n=10):
    x = 1
    for i in range(n):
        yield x
        x += 4
```

```python
def sixth_series(n=10):
    x=4
    for i in range(n):
        yield x
        x += 6
```

```python
def golden_curve(max_n=10):
    fibs = list(fibonacci(max_n))
    fibs.reverse()
    fig, ax = plt.subplots(1)
    last_x, last_y = fibs[0]
    ax.set_xlim(0, last_x + 10)
    ax.set_ylim(0, last_y + 10) 
    plt.axis('off')
    origin = [0, 0]
    p = 0
    arc_points = []
    rects = []
    angle = 90
    for i,(cur_fn, prev_fn) in enumerate(fibs):
        r = random.random()
        b = random.random()
        g = random.random()
        a = random.random()
        colour = (r, g, b, a)
        if i > max_n: break
        arc_radius = cur_fn
        if i in fourth_series(max_n):
            origin[0] = origin[0] + cur_fn
            ax.text(origin[0]+cur_fn/2,origin[1]+cur_fn/2,fibs[i][1],fontsize=(10+fibs[i][0]/2),ha='center',va='center',color='black')
        elif i in sixth_series(max_n):
            origin[1] = origin[1] + cur_fn
            ax.text(origin[0]+cur_fn/2,origin[1]+cur_fn/2,fibs[i][1],fontsize=(10+fibs[i][0]/2),ha='center',va='center',color='black')
        if i%2 == 0:
            cur_fn, prev_fn = prev_fn, cur_fn
            ax.text(origin[0]+cur_fn/2,origin[1]+cur_fn/2,fibs[i][1],fontsize=(10+fibs[i][0]/2),ha='center',va='center',color='black')
        rectangle=Rectangle(origin, prev_fn, cur_fn, angle=0.0,linewidth=2,color=colour)
        #rects.append(rectangle)
        if i == 0: continue
        if i  % 8 == 0:
            p += 1
            continue
        if len(arc_points) == 8: continue
        r1 = rectangle
        coords = [r1.get_xy(), [r1.get_x()+r1.get_width(), r1.get_y()],
                  [r1.get_x()+r1.get_width(), r1.get_y()+r1.get_height()],
                  [r1.get_x(), r1.get_y()+r1.get_height()]]
        arc_points.append((coords[p % 4], arc_radius, angle))
        angle -= 90
        if angle == -360: angle = 0
        p += 3
        ax.add_patch(rectangle)
    for center, radius, angle in arc_points:
        arc = Arc(center, radius*2, radius*2, angle=angle,snap=True,
                  theta1=0, theta2=90.0, edgecolor='black',
                  antialiased=True)
        ax.add_patch(arc)
    #rect_pcs = PatchCollection(rects,alpha=0.1, edgecolor='black',linewidth=2)
    #ax.add_collection(rect_pcs)
    plt.gca().set_aspect("equal")
    plt.show()
```

```python
golden_curve(1)
golden_curve(2)
golden_curve(3)
golden_curve(4)
golden_curve(5)
golden_curve(6)
golden_curve(7)
golden_curve(8)
golden_curve(9)
```
