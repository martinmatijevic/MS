Zadatak 1.: Relativistička kinetička energija

Cilj zadatka je nacrtati graf u kojem se uspoređuje relativistička i obična kinetička energija. <br>
Za precizno računanje relativističke kinetičke energije koristio sam decimal library s preciznošću 50 decimalnih mjesta. <br>
Pri crtanju grafa koristio sam 100 točaka iz intervala $ [ 10^{-2}, 2.99 \cdot  10^{8} ] $.

```python
from scipy.constants import speed_of_light as C
from decimal import Decimal, getcontext
import numpy
import matplotlib.pyplot as plt

%matplotlib inline

getcontext().prec = 50

br_toc=100

def E_rel(m,v):
    return float(Decimal(m)*Decimal(C**2)*(Decimal(1)/(numpy.sqrt(Decimal(1)-(Decimal(v)/Decimal(C))**2))-Decimal(1)))
def E(m,v):
    return float((Decimal(m)*Decimal(v**2))/2)

x=numpy.linspace(1e-2, 2.99e8, br_toc)
y=[0]*br_toc
Y=[0]*br_toc
for i in range(br_toc):
    y[i]=Decimal(x[i])
    y[i]=E_rel(1,y[i])
    Y[i]=Decimal(x[i])
    Y[i]=E(1,x[i])
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, color="red")
plt.plot(x, Y, color="blue", linestyle='-.')
fig = plt.ylim(0, 1e17)
fig = plt.xlim(0, 3e8)
fig = plt.xlabel("v(m/s)", fontsize="x-large")
fig = plt.ylabel("E_kin(J)", fontsize="x-large")
fig = plt.legend(["Erel","E"], loc="best", fontsize="large")
```

Zadatak 2.: Elipsa <br>
Cilj zadatka je nacrtati elipsu, čije je središte u (0,0) te ju zarotirati oko središta za neki proizvoljni kut <br>
te generirati 1000 točaka koje obuhvaćaju pravokutnik oko elipse i nacrtati samo one koje se nalaze unutar nje. <br>
Za rotaciju elipse i pojedinih točaka koristio sam matricu rotacije, a za odabir točaka koje se nalaze unutar elipse sam koristio segmentni oblik elipse dok sam za crtanje elipse koristio parametarski oblik.

```python
import random as rnd
a=10
b=4
fi=numpy.pi/6
br_toc=100
N=1000
t=numpy.linspace(0,2*numpy.pi,br_toc)

Rot=numpy.array([[numpy.cos(fi) , -numpy.sin(fi)],[numpy.sin(fi) , numpy.cos(fi)]])
Ellipse=numpy.array([a*numpy.cos(t), b*numpy.sin(t)])
Rot_Ellipse=numpy.zeros((2,br_toc))
for i in range(br_toc):
    Rot_Ellipse[:,i]=numpy.dot(Rot,Ellipse[:,i])

plt.figure(figsize=(10,10))
plt.plot(Rot_Ellipse[0,:],Rot_Ellipse[1,:], color="green", linewidth=5)
plt.xlim(-1.01*max(a,b),1.01*max(a,b))
plt.ylim(-1.01*max(a,b),1.01*max(a,b))
plt.axhline(0, color='black', alpha=0.5)
plt.axvline(0, color='black', alpha=0.5)

x=[0]*N
y=[0]*N
for i in range (N):
    x[i]=rnd.uniform(-a,a)
    y[i]=rnd.uniform(-b,b)
dots=numpy.array([x,y])
dots_rot=numpy.zeros((2,N))
for i in range(N):
    dots_rot[:,i]=numpy.dot(Rot,dots[:,i])
for i in range (N):
    if ((dots_rot[0,i]*numpy.cos(fi)+dots_rot[1,i]*numpy.sin(fi))**2/a**2+(dots_rot[0,i]*numpy.sin(fi)-dots_rot[1,i]*numpy.cos(fi))**2/b**2)<1:
        plt.scatter(dots_rot[0,i],dots_rot[1,i], marker=".")
```

Zadatak 3.: Monty Hall <br>
Cilj zadatka je izračunati vjerojatnosti dobitka nagrade u Monty Hall kvizu, kod ostajanja pri odabiru i kod mijenjanja odabira početnih vrata. <br>
Pri računanju vjerojatnosti spremao sam brojeve u listu kako bih kasnije izračunao srednju vrijednost, i koristio sam brojač za pojedina vrata <br>
koja sudionik kviza otvori kako bih na kraju mogao podijeliti vrijednost brojača s brojem semplova za određivanje vjerojatnosti. <br>
Također, potrebno je izračunati nepouzdanost rezultata, za što sam koristio formulu za standardnu devijaciju aritmetičke sredine.

```python
a=[-1,0,1]
N=0
n=10000000
switch=0
stay=0
sigma=0
array=[0]*n
while N<n:
    x=rnd.choice(a)
    array[N]=x
    if x==0:
        stay+=1
    if x==1:
        switch+=1
    if x==-1:
        sigma+=1
    N+=1
x=(stay*0+switch*1+sigma*(-1))/n
N=0
switch=switch+sigma
sigma=0
while N<n:
    sigma=sigma+(array[N]-x)**2
    N+=1
print("Vjerojatnost dobitka kod ostajanja pri odabiru je: %.4f%%" % (stay/n*100))
print("Vjerojatnost dobitka kod mijenjanja odabira je: %.4f%%"% (switch/n*100))
print("Nepouzdanost rezultata je: %.7f" % (numpy.sqrt(sigma/(n*(n-1)))))
```
